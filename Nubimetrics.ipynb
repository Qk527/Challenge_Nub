```python
import requests
import json
import pandas as pd
```

```python
def main_get (categoria):
    url = "https://api.mercadolibre.com/sites/MLA/search?category="+categoria+"&+limit="+str(limite)+"#json"
    response = requests.get(url)
    if response.status_code == 200:
        respo_json = json.loads(response.text)
    else:
        print ("Something is wrong")
    return respo_json['results']
```

```python
limite = 100
elec_aud_vid_data = main_get("MLA1000")
```

```python
print (json.dumps(elec_aud_vid_data, indent=4))
```

```python
from datetime import datetime
from datetime import timedelta

def generate_path_mult (date_u,days_u):
    date_obj = datetime.strptime(date_u, '%Y/%m/%d')
    for i in range(days_u, -1, -1):
        date_calc = date_obj - timedelta(days=i)
        result = "https://importantdata@location/"+str(date_calc.year)+"/"+str(date_calc.month)+"/"+str(date_calc.day)+"/"
        print (result)
    return result
```

```python
print(generate_path_mult("2020/12/12",10))
```
